```python
import numpy as np
import matplotlib.pyplot as plt
 
plt.figure(figsize=(10, 6))
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
 
plt.plot(x, y, label='sin(x)', color='green')
plt.plot(x, z, label='cos(x)', color='orange')
plt.title('  sin(x)  cos(x)')
plt.xlabel('x')
plt.ylabel(' ')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=20, color='black', alpha=0.7)
plt.title('  ')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.show()
```

```python
matrix = np.random.randint(1, 11, size=(5, 5))
print(" 5x5:")
print(matrix)
print("\n :", np.mean(matrix))
print(" :", np.max(matrix))
print(" :", np.min(matrix))
print("  :", np.sum(matrix, axis=0))
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(matrix, cmap='viridis')
plt.colorbar()
plt.title('  ')
 
#  
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, matrix[i, j], ha='center', va='center', color='white')
plt.show()
```
