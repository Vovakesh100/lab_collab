```python
import matplotlib.pyplot as plt

data = {"Еда": 45, "Транспорт": 30, "Развлечения": 25}

#размеры
plt.figure(figsize=(8, 6))

# Создаем диаграмму с процентами
plt.pie(data.values(),
        labels=None,  # yбираем, так как будет легенда
        autopct='%1.0f%%',  # целое число
        colors=['#ff9999', '#66b3ff', '#99ff99'])

# Добавляем легенду
plt.legend(data.keys(),
           title="Категории",
           loc="center left",
           bbox_to_anchor=(1, 0.5))

# Добавляем заголовок
plt.title("Распределение моих расходов", pad=20)

# Сохраняем и показываем
plt.tight_layout()  # Чтобы не обрезалось
plt.savefig('expenses.png', dpi=100, bbox_inches='tight')
plt.show()
```
